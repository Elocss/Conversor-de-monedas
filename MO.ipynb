```python
from datetime import datetime

def calcular_tarifa(km):
    valor_km = 400
    impuesto = 0.10
    total = km * valor_km

    # Lógica de hora
    hora = datetime.now().hour
    alta_demanda = 0

    if (16 <= hora <= 18) or (0 <= hora <= 5):
        alta_demanda = 0.10
        print("Aplicando tarifa de alta demanda (+10%)")

    total_final = total + (total * impuesto) + (total * alta_demanda)
    return total_final

# Ejemplo de uso
distancia = 10 # 10 km recorridos
print(f"El total a pagar es: ${calcular_tarifa(distancia)} ARS")
```

```python
from kivymd.app import MDApp
from kivy.lang import Builder
from kivy.uix.boxlayout import BoxLayout
from kivy.uix.image import Image # Import Image widget for FitImage
from kivy.uix.scrollview import ScrollView # Import ScrollView for MDList

 # Definir MapView como un marcador de posición para la ejecución en Colab si no es un widget estándar de Kivy
# Si MapView es un widget personalizado o de una librería específica, debe importarse o simularse
# Por simplicidad, crearemos un marcador si no es un widget estándar de Kivy
# Si MapView es parte de una librería que necesitas, deberías instalarla e importarla.
# Por ahora, asumamos que podría ser un widget personalizado o un marcador de posición.
class MapView(BoxLayout):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.add_widget(Image(source='https://www.gstatic.com/earth/applications/globe/v2/globe_preview_dark.png')) # Placeholder image

class MooApp(MDApp):
    def build(self):
        # Load the KV file directly
        self.theme_cls.primary_palette = "Blue"
        return Builder.load_file('moo.kv')

    # Define the methods called in the KV file, even if they are empty for now
    def ir_a_perfil(self):
        print("Navigating to profile!")

    def explicar_sistema(self):
        print("Explaining the system!")

# To run the app, you would typically call:
# MooApp().run()
# However, running Kivy apps directly in Colab without a display server is not straightforward.
# The above code sets up the app, but will not display the GUI in the Colab output.
print("KivyMD app setup complete. To run this app with a GUI, execute MooApp().run() in a local environment with a display server.")
```
